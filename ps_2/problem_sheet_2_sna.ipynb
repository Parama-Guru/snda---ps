```python
node_d=G_d.number_of_nodes()
print(node_d)
print(dict(G_d.degree()))
eccentricity_d=nx.eccentricity(G_d)
print(eccentricity_d)
degree_centrality_d=nx.degree_centrality(G_d)
max_degree_centrality_d=max(degree_centrality_d,key=degree_centrality_d.get)
print("the node with maximum degree centrality:",max_degree_centrality_d,degree_centrality_d[max_degree_centrality_d])
print(degree_centrality_d)
closeness_centrality_d=nx.closeness_centrality(G_d)
max_closeness_centrality_d=max(closeness_centrality_d,key=closeness_centrality_d.get)
print("the node with maximum closeeness centrality:",max_closeness_centrality_d,closeness_centrality_d[max_closeness_centrality_d])
betweenness_centrality_d=nx.betweenness_centrality(G_d)
max_betweenness_centrality_d=max(betweenness_centrality_d,key=betweenness_centrality_d.get)
print("the node with the maximum betweeness centrality is",max_betweenness_centrality_d,betweenness_centrality_d[max_betweenness_centrality_d])
eigenvector_centrality_d=nx.eigenvector_centrality(G_d)
max_eigenvector_centrality_d=max(eigenvector_centrality_d,key=eigenvector_centrality_d.get)
print("the node with the higghest eigen vector centrality is:",max_eigenvector_centrality_d,eigenvector_centrality_d[max_eigenvector_centrality_d])
```

```python
node_n=G_n.number_of_nodes()
print(node_n)
print(dict(G_n.degree()))
if nx.is_connected(G_n):
    eccentricity_n = nx.eccentricity(G_n)
else:
    # Get all connected components
    components = (G_n.subgraph(c).copy() for c in nx.connected_components(G_n))

    eccentricity_n = {}
    for component in components:
        ecc = nx.eccentricity(component)
        eccentricity_n.update(ecc)
print(eccentricity_n)
degree_centrality_n=nx.degree_centrality(G_n)
max_negree_centrality_n=max(degree_centrality_n,key=degree_centrality_n.get)
print("the node with maximum degree centrality:",max_negree_centrality_n,degree_centrality_n[max_negree_centrality_n])
closeness_centrality_n=nx.closeness_centrality(G_n)
max_closeness_centrality_n=max(closeness_centrality_n,key=closeness_centrality_n.get)
print("the node with maximum closeeness centrality:",max_closeness_centrality_n,closeness_centrality_n[max_closeness_centrality_n])
betweenness_centrality_n=nx.betweenness_centrality(G_n)
max_betweenness_centrality_n=max(betweenness_centrality_n,key=betweenness_centrality_n.get)
print("the node with the maximum betweeness centrality is",max_betweenness_centrality_n,betweenness_centrality_n[max_betweenness_centrality_n])
eigenvector_centrality_n=nx.eigenvector_centrality(G_n)
max_eigenvector_centrality_n=max(eigenvector_centrality_n,key=eigenvector_centrality_n.get)
print("the node with the higghest eigen vector centrality is:",max_eigenvector_centrality_n,eigenvector_centrality_n[max_eigenvector_centrality_n])
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a sample graph
G = nx.erdos_renyi_graph(10, 0.3)

# Define layouts
layouts = {
    "spring": nx.spring_layout,
    "circular": nx.circular_layout,
    "random": nx.random_layout,
    "shell": nx.shell_layout,
    "spectral": nx.spectral_layout,
}

# Plot graphs with different layouts
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
axes = axes.flatten()

for ax, (name, layout) in zip(axes, layouts.items()):
    pos = layout(G_d)
    nx.draw(G_d, pos, ax=ax, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray')
    ax.set_title(f"{name} layout")

plt.tight_layout()
plt.show()
```
